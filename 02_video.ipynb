```python
## video 파일 입출력
# 비디오의 재생은 연속된 이미지를 재생하는 것
# 프레임: 연속된 이미지 중 하나
# 비디오를 재생하기 위해서는 프레임을 연속적으로 재생

```

```python
import cv2
video_path='videos/a.mp4'

```

```python
cap=cv2.VideoCapture(video_path)
print(cap.isOpened())    # 열렸는지 확인


while cap.isOpened():
    ret, frame = cap.read()    # 파일을 읽어서 프레임을 반환(ret: 프레임이 있는지 체크)
    print(frame)
    
    if not ret:
        break

    cv2.imshow('video',frame)

    if cv2.waitKey(1)==ord('q'):
        break
cap.release()    # 영상 객체를 해제
cv2.destroyAllWindows()
```

```python
## 영상의 프레임 불러오기
# VideoCapture.get(cv2.CAP_PROP_FPS)  영상의 정보 불러오기

cap=cv2.VideoCapture(video_path)
fps=cap.get(cv2.CAP_PROP_FPS)

print(fps)


while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break
    cv2.imshow('video',frame)

    if cv2.waitKey(int(1000/fps)) == ord('q'):             # 1초에 프레임이 몇개인가 프레임 하나당의 초
        break

cap.release()
cv2.destroyAllWindows()
```

```python
## 카메라 입출력
cap=cv2.VideoCapture(0)    # 0번의 카메라를 불러온다
print(cap.isOpened())    # 카메라 확인

if not cap.isOpened():
    exit()

while True:
    ret, frame=cap.read()

    if not ret:
        break

    cv2.imshow('camera', frame)

    if cv2.waitKey(1)==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 비디오 쓰기
cap=cv2.VideoCapture(video_path)
fourcc=cv2.VideoWriter_fourcc(*'DIVX')    # *붙이면 글자마다 한칸씩 떨어진다.
fps=cap.get(cv2.CAP_PROP_FPS)
width=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

out=cv2.VideoWriter('videos/b.mp4', fourcc, fps, (round(width), round(height)))   # fourcc: 코덱

while cap.isOpened():
    ret, frame=cap.read()

    if not ret:
        break

    out.write(frame)    # 소리는 들어가지 않음
    cv2.imshow('camera',frame)

    if cv2.waitKey(int(1000/fps))==ord('q'):
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```
