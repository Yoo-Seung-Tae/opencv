```python
## 히스토그램: openCV에서는 이미지에 존재하는 픽셀의 밝기를 표현하는데 사용함

import cv2
import matplotlib.pyplot as plt

img=cv2.imread('images/iu.jpg')

# 채널: 0은 블루, 1은 그린, 2는 레드
hist_b=cv2.calcHist([img], [0], None, [256], [0,256])      # 리스트 형태로 이미지 넣기, 채널, 마스크, x축 구간(0~255), y축 구간(0이상 256미만)
hist_g=cv2.calcHist([img], [1], None, [256], [0,256])      # 리스트 형태로 이미지 넣기, 채널, 마스크, x축 구간(0~255), y축 구간(0이상 256미만)
hist_r=cv2.calcHist([img], [2], None, [256], [0,256])      # 리스트 형태로 이미지 넣기, 채널, 마스크, x축 구간(0~255), y축 구간(0이상 256미만)

plt.plot(hist_b, color='b')
plt.plot(hist_g, color='g')
plt.plot(hist_r, color='r')
plt.title('histogram')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.show()
```
