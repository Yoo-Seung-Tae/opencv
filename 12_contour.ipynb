```python
## 윤곽선 검출: 경계선을 연결한 선
# 흑백 이미지 -> 이진화 -> 윤곽선 찾기(findcontour) -> 윤곽선 그리기(drawcontour)

import cv2

img=cv2.imread('images/iu.jpg')

# 윤곽선 검출은 원본에 영향을 주므로 미리 깊은 복사
coppied=img.copy()
coppied_2=img.copy()
```

```python
# 그레이 스케일
gray=cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)
gray_2=cv2.cvtColor(coppied_2, cv2.COLOR_BGR2GRAY)

# 이진화(오츠)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
ret_2, binary_2 = cv2.threshold(gray_2, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# 윤곽선 찾기
contours, hierachy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)          # 전체 윤곽선 검출
contours_2, hierachy_2 = cv2.findContours(binary_2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)     # 외부 윤곽선 검출

# 윤곽선 그리기
cv2.drawContours(coppied, contours, -1, (255,0,0), 2) 
cv2.drawContours(coppied_2, contours_2, -1, (255,0,0), 2) 

cv2.imshow('1', img)
cv2.imshow('2', coppied)
cv2.imshow('3', coppied_2)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 윤곽선 경계 사격형(bounding rectangle): 윤곽선을 둘러싼 사격형 그리기

# 그레이 스케일
gray=cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)

# 이진화(오츠)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# 윤곽선 찾기
contours, hierachy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)          # 전체 윤곽선 검출

# 윤곽선 사격형 그리기
for contour in contours:
    x, y, width, height = cv2.boundingRect(contour)
    cv2.rectangle(coppied, (x,y), (x+width,y+height),(255,0,0), 2)
    print(x, y)


cv2.imshow('1', img)
cv2.imshow('2', coppied)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## 실습 : 윤곽선 경계 사격형 + 트랙바
import cv2
img=cv2.imread('images/vehicles.png')

# 윤곽선 검출은 원본에 영향을 주므로 미리 깊은 복사
coppied=img.copy()

# 그레이 스케일
gray=cv2.cvtColor(coppied, cv2.COLOR_BGR2GRAY)

# 이진화(오츠)
ret, binary = cv2.threshold(gray, -1, 120, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# 윤곽선 찾기
contours, hierachy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)          # 전체 윤곽선 검출

name='Trackbar'
cv2.namedWindow(name)
cv2.createTrackbar('Trackbar', name, 0, len(contours)-1, lambda x:x)           # 트랙바 창 이름, 트랙바 조정 값 이름, 시작점, 최대점, 수식(무의미해도 없으면 안됨, 변동이 있을 시 변환)

# 윤곽선 사격형 그리기
while True:
    threshold=cv2.getTrackbarPos('Trackbar', name)
    coppied_2=coppied.copy()
    x, y, width, height = cv2.boundingRect(contours[threshold])


    cv2.rectangle(coppied_2, (x,y), (x+width,y+height),(255,0,0), 2)
    cv2.imshow(name, coppied_2)


    if cv2.waitKey(1) == ord('q'):
        break


cv2.destroyAllWindows()
```
